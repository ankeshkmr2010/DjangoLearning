```python
import cv2
import numpy as np

cv2.__version__

# popular image processing library https://github.com/python-pillow/Pillow
```

```python
img = cv2.imread('16.jpg')
plt.imshow(img)
```

```python
img.shape
```

# Lets do some geometric transformation

```python
res = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_CUBIC)
plt.imshow(res)
```

```python
height, width = img.shape[:2]
res = cv2.resize(img,(2*width, 2*height), interpolation = cv2.INTER_CUBIC)
plt.imshow(res)
```

```python
# Translation
# Translation is the shifting of objects location.
# If you know the shift in (x,y) direction, let it be (t_x,t_y), you can create the
# transformation matrix \textbf{M} as follows:
img = cv2.imread('16.jpg',0)
rows,cols = img.shape

M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))
plt.imshow( dst)
```

```python
rows,cols = img.shape

M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
dst = cv2.warpAffine(img,M,(cols,rows))
plt.imshow( dst)
```

```python
# Image smoothing is an important image processing technique that performs
# blurring and noise filtering in an image.
# The famous smoothing algorithms in use are Blur, Gaussian Blur, Median Blur and Bilateral Filter.
# Apply blur
img = cv2.imread('16.jpg')
img1 = cv2.blur(img,(3,3))
# display the images
plt.imshow(img)
plt.imshow(img1)
```
